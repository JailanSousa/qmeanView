```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bokeh.models import ColumnDataSource, CustomJS, HoverTool
from bokeh.plotting import figure, show
from scipy.stats import gaussian_kde

from bokeh.io import output_notebook

output_notebook()

%matplotlib inline
```

```python
df = pd.read_csv('qmean_reference/qmean_refrence.csv')
qmean_ref = df[df['qmean4_norm'] > 0.5].sort_values(by='protein_size')
models = pd.read_csv("models_fix.csv")
```

```python
s = ColumnDataSource(models)

p = figure(width=400, height=400, tools="lasso_select", title="Select Here",  y_range=(0, 1.5))
p.scatter('protein_size', 'qmean4_norm', color='navy', size=8, source=s, alpha=0.4,
          selection_color="firebrick")

s2 = ColumnDataSource(data=dict(x=[0, 100], ym=[0.5, 0.5]))
s3 = ColumnDataSource(data=dict(x=[0, 100], ym=[0.7, 0.7]))
p.line(x='x', y='ym', color="orange", line_width=5, alpha=0.6, source=s2)
p.line(x='x', y='ym', color="orange", line_width=5, alpha=0.6, source=s3)
show(p)
```

```python
x = qmean_ref['protein_size']
y = qmean_ref['qmean4_norm']
z = qmean_ref['qmean4_zscore']
```

```python
def kde(x, y, N):
    xmin, xmax = x.min(), x.max()
    ymin, ymax = y.min(), y.max()

    X, Y = np.mgrid[xmin:xmax:N*1j, ymin:ymax:N*1j]
    positions = np.vstack([X.ravel(), Y.ravel()])
    values = np.vstack([x, y])
    kernel = gaussian_kde(values)
    Z = np.reshape(kernel(positions).T, X.shape)

    return X, Y, Z

x, y, z = kde(x, y, 300)

p = figure(height=400, x_axis_label="hp", y_axis_label="mpg",
           background_fill_color="#fafafa", tools="", toolbar_location=None,
           title="Kernel density estimation plot of HP vs MPG")
p.grid.level = "overlay"
p.grid.grid_line_color = "black"
p.grid.grid_line_alpha = 0.05

palette = Blues9[::-1]
levels = np.linspace(np.min(z), np.max(z), 10)
p.contour(x, y, z, levels[1:], fill_color=palette, line_color=palette)

show(p)
```
